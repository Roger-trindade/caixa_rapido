# Iris

```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('Iris.csv')
df.sample(5)
```

```python
df2 = df.copy()
df2
```

```python
df2.drop(['Id', 'Species'], axis = 1, inplace = True)
df2
```

```python
df2.rename(
    {'SepalLengthCm' : 'SeComprimento',
     'SepalWidthCm' : 'SeLargura',
     'PetalLengthCm' : 'PeComprimento',
     'PetalWidthCm' : 'PeLargura'},
    axis = 1, inplace = True
)

df2
```

# Criando grafico

```python
plt.scatter(df2['SeComprimento'], df2['SeLargura'])
```

```python
plt.scatter(df2['PeComprimento'], df2['PeLargura'])
```

# KMeans

```python
modelo = KMeans(n_clusters = 3, random_state = 93)
modelo
```

```python
modelo.fit(df2)
```

```python
modelo.labels_
```

```python
df['Previsto'] = modelo.labels_
df
```

```python
df['Species'] = df['Species'].replace({ 'Iris-setosa' : 1,
                                        'Iris-versicolor' : 0,
                                        'Iris-virginica' : 2})

df
```

```python
numero_de_acerto = sum(df['Species'] == df['Previsto'])
numero_de_acerto
```

```python
print(f'O nosso modelo teve uma taxa de acerto de {(numero_de_acerto / df.shape[0]) * 100}%')
```

# ELbow method

```python
inercia = []

for valor in range(1, 10):
  modelo = KMeans(n_clusters = valor)
  modelo.fit(df2)
  inercia.append(modelo.inertia_)

inercia
```

```python
plt.plot(range(1, 10), inercia,
         marker = 'o',
         linestyle = '--',
         color = 'black')
```

```python
import pandas as pd
import numpy as np

from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, accuracy_score
```

# Modelo Ã³timo

## Carregamento dos dados

```python
data = {
    'Marketing' : [100, 200, 300, 400, 500, 600, 700, 800, 900, 1000],
    'Vendas' : [50, 100, 150, 200, 250, 300, 350, 400, 450, 500]
}
df = pd.DataFrame(data)
df
```

Espera-se que a funÃ§Ã£o seja do tipo y = 0.5 * x

## SeparaÃ§Ã£o X e Y

```python
X = df[['Marketing']]
y = df['Vendas']
```

```python
X
```

```python
y
```

## SeparaÃ§Ã£o treino e teste

```python
X_treino, X_teste, y_treino, y_teste = train_test_split(X, y, test_size = 0.3, random_state = 93)
```

```python
X_treino
```

```python
X_teste
```

## Criar o modelo

```python
modelo = LinearRegression()
```

```python
modelo
```

```python
modelo.fit(X_treino, y_treino)
```

```python
modelo.coef_
```

```python
modelo.intercept_ # Ã o mesmo de -0.00000000000001 e Ã© praticamente 0
```

```python
previsao = modelo.predict(X_teste)
previsao
```

```python
X_teste
```

## Calculando erro do modelo

```python
mse = mean_squared_error(y_teste, previsao)
mse
```

# modelo - Classificador

## Carregamento de dados

```python
df = pd.DataFrame({
    'Idade' : [25, 35, 45, 20, 30, 50, 40, 55, 60, 70],
    'Renda' : [3000, 4000, 5000, 2000, 3500, 6000, 5500, 7000, 8000, 9000],
    'Compra' : [0,0,1,0,1,1,1,1,1,1]
})
df
```

```python
X = df[['Idade', 'Renda']]
y = df['Compra']
```

```python
X
```

## SeparaÃ§ao treino teste

```python
X_treino, X_teste, y_treino, y_teste = train_test_split(X, y, test_size = 0.3, random_state = 93)
```

```python
X_treino
```

## Criando o modelo

```python
modelo_logistico = LogisticRegression()
modelo_logistico
```

```python
modelo_logistico.fit(X_treino, y_treino)
```

```python
modelo_logistico.coef_
```

```python
modelo_logistico.intercept_
```

```python
previsao_logistica = modelo_logistico.predict(X_teste)
previsao_logistica
```

```python
X_teste
```

## Calcula acurÃ¡cia

```python
acuracia = accuracy_score(y_teste, previsao_logistica)
acuracia
```
